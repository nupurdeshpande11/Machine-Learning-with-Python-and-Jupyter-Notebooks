```python
sns.pairplot(nupur,kind="reg",vars=['STA','Snowfall','MinTemp','MO'])
#This pairplot is designed to understand the nature of snowfall during world war2.Following are observations:
#1. STAs numbering around 12500 to 16000 experienced maximum snow at more than 80mm.There is a linerally regressive(reducing)
# relation between the STAs and Snowfall.Lower number of STAs experience more Snowfall.
#2.Snowfall and Month have a relationship that is linerally reducing dfrom month 4 to month 10. Snowfall is at its peak during 
# the December and then through April its is at its highest. These are the colder months in Europe.
#3.There is a lineraly inverse relationship between MinTemp and Snowfall. As the MinTemp drops snowfall increases a lot. 
# Outliers: From the SnowfallvsMinTemp graph we can see that around tempeatures just around 0 very few times snow greater than 75
# mm has fallen. From the graph of STA vs MinTemp we see that very few(outlier)STAs 12000-16000 have recorded subzero temperature
```

```python
sns.pairplot(nupur,vars=['WindGustSpd','MaxTemp','MinTemp'],kind="reg")
#Want to analyse around which temeperatures WindGustSpeed is at its peak and the linear relationship between Min and Max Temp.
#We can analyse that WindGusts occur at temperatures above 25 degrees. There is little correlation between temperature and Wind
#GustSpeed. The graph of MinTemp to WindGustSpeed increases with little slope.The graph of MaxTemp to WGS has very little slope.
#Outliers: windgustspeeds taking place below temperatures of 20 and close to 40 are the main outliers. 
```

```python
sns.pairplot(nupur,vars=['WindGustSpd','DR','SPD','PGT'],kind="reg")
# In this plot we compare WindGustSpeed with it speed,angle and peak gust time.
#There is a increasing linear relationship between WindGustSpeed and Speed such that as WindGust becomes stronger its speed
# increases. There are no outliers. 
# DR(Direction) and WindGust are also linear relation but have less correlation and many outliers. This is typical since the 
#speed of wind does not always determine it's direction. Higher speed(75 kmph) winds have the same direction as 40kmph wind
# As the WindGust speed increases the Peak Gust(PGT) ie time for which a blast of wind occurs increases. Thus PGT and WindSpdGust
# have a increasing linear relationship. Stronger WSG the more time PGT occurs. 
#There are many outliers since PGT is a measurement of time. Gust of same speed can stay for a few seconds and later stay for a 
# few minutes depending on other weather conditions. 
```

```python
sns.pairplot(nupur,vars=['WindGustSpd','Precip','Snowfall','MO','YR'],kind="reg")
#To check relationship betwwen WGS,Precip,Snowfall and the months they should occcur. Also analyse how many years have got how 
#much of the former three. Also check any relationship between WGS,Precip and Snowfall
# WGS: WGS has a linerally increasing relationship 1944 onwards. Most WGS have occured during 1945. WGS share a linearly decrease
# relationship with months such that January has highest WGS and August has lowest. Other months have WGS on average. 
# Precipitation and Month and Year: Precipitation limearly increases upto year 1944. Lesser rain is seen in 46. Precip has a
#linearlly increasing relationship with months too. Months 6 to 11 see highest occurance of precip. 
#Snowfall: Linearly decreasing relationship with months. Highest Snowfall in December ,Jan to Apri. Most snow between years 44-46
#WGS and Precip: no relationship other than few outliers, days that have been windy and rained.
#WGS and Snowfall: have no relationship. Completely uncorrelated
#Snowfall and Precip:Again snowfall and precip have no relationship only a few outliers. 
```
